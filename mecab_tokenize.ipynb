## 辞書を変えながら形態素解析

```
import MeCab
sample_txt = "彼女は子どもと竈門炭治郎のコスプレをした。"

# /var/lib/mecab/dic/debian/sys.dic
m = MeCab.Tagger()
print("Mecab:\n", m.parse(sample_txt))

# IPA 辞書：標準ではないが、MeCab と共によく用いられる。
m = MeCab.Tagger("-d /var/lib/mecab/dic/ipadic-utf8")
print("Mecab IPADIC:\n", m.parse(sample_txt))

# 代替表記が便利
m = MeCab.Tagger("-d /var/lib/mecab/dic/juman-utf8")
print("Mecab JUMAN:\n", m.parse(sample_txt))

# NEologd: 新語などに強い
m = MeCab.Tagger("-d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd")
print("Mecab ipadic NEologd:\n",m.parse(sample_txt))

# Unidic 書：Mecab での利用が推奨されている。
import unidic
# print(unidic.DICDIR)
m = MeCab.Tagger('-d ' + unidic.DICDIR)
print("Mecab Unidic:\n", m.parse(sample_txt))
```

```
# NEologd では「竈門炭治郎」は認識できたが、「炭治郎」は？
sample_txt = "彼女は子どもと炭治郎のコスプレをした。"

m = MeCab.Tagger("-d /usr/lib/x86_64-linux-gnu/mecab/dic/mecab-ipadic-neologd")
print("Mecab ipadic NEologd:\n",m.parse(sample_txt))
```

## 品詞ごとの処理の仕方

```
# 単純に表示するなら m.parse() でよいが、各要素を処理したい場合は、以下のようにする。
node = m.parseToNode(sample_txt)

result = [] # 名詞だけ抽出
while node:
  feature = node.feature.split(',')
  if node.feature.split(',')[0] == '名詞': # 名詞の時だけ出力
    #print(node.surface, feature[0]) # feature の例 ['名詞', '代名詞', '一般', '*', '*', '*', '彼女', 'カノジョ', 'カノジョ']
    result.append(node.surface)
  node = node.next # 次のノードへ移動

print(result)
```

```
# JSON 形式で、名詞のリストをファイルに保存
import json
json_filename = "test.json"

with open(json_filename, 'w') as out:
  json.dump(result, out) 

with open(json_filename, "r") as f:
  read_data = json.load(f)
  print(read_data)
  print(read_data[1])
  print(type(read_data))
```
